```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 10000)
def relu(x):
    return np.maximum(0, x)

y_relu = relu(x)

plt.figure(figsize=(10, 8))
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True, which='both')
plt.plot(x, y_relu)

plt.savefig('ReLU.png')
plt.savefig('ReLU.eps')
```

```python
def prelu(x, alpha):
    if x < 0:
        return alpha * x
    else:
        return x
    
vprelu = np.vectorize(prelu)
```

```python
y_prelu = vprelu(x, 0.1)

plt.figure(figsize=(10, 8))
plt.xlabel('x')
plt.ylabel('PReLU(x)')
plt.grid(True, which='both')
plt.plot(x, y_prelu)

plt.savefig('PReLU.png')
plt.savefig('PReLU.eps')
```
